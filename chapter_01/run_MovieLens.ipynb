```python
import pandas as pd
```

```python
unames = ['user_id','gender','age','occupation','zip']

users = pd.read_table('/Users/yifanwang/Documents/data_mining/pydata-book/ch02/movielens/users.dat', sep='::', header=None, names=unames)

rnames = ['user_id', 'movies','rating', 'timestamp']

ratings = pd.read_table('/Users/yifanwang/Documents/data_mining/pydata-book/ch02/movielens/ratings.dat',  sep='::', header=None, names=rnames)

mnames = ['movie_id','title','genres']

movies = pd.read_table('/Users/yifanwang/Documents/data_mining/pydata-book/ch02/movielens/movies.dat',  sep='::', header=None, names=mnames, encoding='latin-1')
```

```python
print(users.head())
print(ratings.head())
print(movies.head())
```

```python
data = pd.merge(pd.merge(users, ratings, on='user_id'), movies, left_on = 'movies', right_on='movie_id')
data
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title.head()
```

```python
ratings_by_title.index
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
# mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
top_female_ratings[:10]
```
